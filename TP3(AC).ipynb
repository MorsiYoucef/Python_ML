```python
import matplotlib.pyplot as plt

# Données
x = [0, 5, 10, 15, 20, 25, 28]  # Rang de l'année
y = [21.9, 26.4, 36.9, 44.7, 67.0, 82.0, 97.9]  # Nombre de passagers (en millions)

# Création du graphique
plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Nombre de passagers")
plt.title("Fréquentation des lignes aériennes en fonction du rang de l'année", fontsize=14)
plt.xlabel("Rang de l'année ($x_i$)", fontsize=12)
plt.ylabel("Nombre de passagers ($y_i$) en millions", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

### ajustement exponential d'apres l'augmentation des nombres passagers au fil des annees

```python
import numpy as np

# Conversion des données en arrays pour les calculs
x_array = np.array(x)
y_array = np.array(y)

# Moyennes de x et y
x_mean = np.mean(x_array)
y_mean = np.mean(y_array)

# Calcul des coefficients de la droite de régression
cov_xy = np.sum((x_array - x_mean) * (y_array - y_mean))  # Covariance entre x et y
var_x = np.sum((x_array - x_mean)**2)  # Variance de x
a = cov_xy / var_x  # Pente de la droite
b = y_mean - a * x_mean  # Ordonnée à l'origine

# Calcul du coefficient de corrélation linéaire
var_y = np.sum((y_array - y_mean)**2)  # Variance de y
r = cov_xy / np.sqrt(var_x * var_y)  # Coefficient de corrélation

a, b, r
```

```python
# Calcul de z = ln(y)
z_array = np.log(y_array)

# Moyennes de x et z
z_mean = np.mean(z_array)

# Calcul des coefficients de la droite de régression pour z = c * x + d
cov_xz = np.sum((x_array - x_mean) * (z_array - z_mean))  # Covariance entre x et z
c = cov_xz / var_x  # Pente
d = z_mean - c * x_mean  # Ordonnée à l'origine

# Expression de y en fonction de x : y = e^d * e^(c * x)
e_d = np.exp(d)

# Estimation pour 2011 (x = 31)
















x_2011 = 31
y_2011 = e_d * np.exp(c * x_2011)

# Calcul de l'augmentation prévue par le modèle entre 2008 et 2011
y_2008 = y_array[-1]  # Dernière valeur des données (2008)
augmentation_model = ((y_2011 - y_2008) / y_2008) * 100  # En pourcentage

c, d, e_d, y_2011, augmentation_model
```
