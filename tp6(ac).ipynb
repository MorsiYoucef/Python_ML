```python
employee = pd.read_csv('employees.csv')
employee.head()
```

```python
employee.isnull().sum()
```

```python
employee['Gender'] = employee['Gender'].fillna(method='ffill')
```

```python
print(employee)
```

```python
employee.isnull().sum()
```

```python
most_frequent_senior = employee['Senior Management'].mode()[0]  # Get the most frequent value
employee['Senior Management'] = employee['Senior Management'].fillna(most_frequent_senior)
```

```python
employee.isnull().sum()
```

```python
employee['Team'] = employee['Team'].fillna(method='ffill')
```

```python
employee.isnull().sum()
```

```python
employee.drop(columns="First Name")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
employee['Gender'] = encoder.fit_transform(employee['Gender'])
employee['Salary'] = encoder.fit_transform(employee['Salary'])
employee['Senior Management'] = encoder.fit_transform(employee['Senior Management'])
employee['Team'] = encoder.fit_transform(employee['Team'])


# Visualiser les valeurs manquantes
plt.figure(figsize=(10, 6))
sns.heatmap(employee, cbar=False, cmap='viridis')
plt.title("Données manquantes")
plt.show()
```
