```python
import numpy as np
import pandas as pd
#from scipy import stats as st
from sklearn import datasets
iris = datasets.load_iris()
#print(dir(iris))#['DESCR', 'data', 'feature_names', 'filename', 'target']
iris_df = pd.DataFrame(iris.data) # cr√©er un dataframe
#print(iris_df .info())
columns = iris.feature_names
iris_df.columns = columns
print(iris_df.corr()) # Pearson correlation test
print(iris_df.cov()) # covariance
iris_df.corr('spearman') # Spearman Rank correlation
```
